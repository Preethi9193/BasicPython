```python
df1 = sns.load_dataset('tips')
df1.head()
```

```python
sns.boxplot(data =df1)
```

```python
sns.boxplot(x='tip',data = df1)
```

```python
sns.boxplot(x='tip',y='sex',data = df1,hue = 'smoker')
```

```python
### to apply the themes to plot
sns.set_theme('poster')
sns.boxplot(x='tip',y='sex',data = df1,hue = 'smoker')
```

```python
## to apply colors

sns.set_theme('talk')
sns.boxplot(x='tip',y='sex',data = df1,hue = 'smoker',palette = 'Set3')
```

```python
sns.set_theme('paper')
sns.boxplot(x='total_bill',y='sex',data = df1,hue = 'day',palette = 'Set3',linewidth = 6)
```

```python
### to get the boxplot in horizontally
sns.boxplot(data = df1, orient = 'h',palette = 'Set2')
```

### Kde plots
- Kernel density estimator plot
- sns.kdeplot()
- we can distributed data very smoothly based on the density  
- we can plot univariate or bivariate distributions


```python
sns.set_theme('notebook')
sns.kdeplot(data = df1)
```

```python
sns.set_theme('talk')
sns.kdeplot(x='tip',data = df1)
```

```python
sns.kdeplot(x='tip',y = 'total_bill',data = df1)
```

```python
sns.kdeplot(x='tip',y = 'total_bill',data = df1,hue = 'sex')  
### it generate a spiral type output when we have two integer type  as input
```

```python
 sns.kdeplot(x='tip',data = df1,bw_adjust=0.5)  ## less smoothness
```

```python
sns.kdeplot(x='tip',data = df1,bw_adjust=5,hue = 'day')
```

```python
### when we have multiple parameters distributions
## we can use stack to fill the distributions
sns.kdeplot(x='tip',data = df1,bw_adjust=5,hue = 'day',multiple = 'stack')
```

```python
### bar plot

sns.barplot(data =df1)
```

### point plot

- to pointout data or values 
- sns.pointplot()
- default it generate scatter plot and shows point estimation between two kinds of intervals

```python
sns.pointplot(data = df1)
```

```python
sns.pointplot(x='total_bill',y = 'day',data = df1)
```

```python
sns.pointplot(x='total_bill',y = 'day',data = df1,hue = 'sex')
```

```python
#### to seperate the interval lines dodge
sns.pointplot(x='total_bill',y = 'day',data = df1,hue = 'sex',dodge = True)
```

```python
sns.set_theme(style = 'whitegrid')
sns.pointplot(x='total_bill',y = 'day',data = df1,dodge = True,estimator = np.mean)
```

### joint plot

- it joins the two plots
- sns.jointplot()

```python
sns.set_theme(style = 'darkgrid')
sns.jointplot(df1['day'])
```

```python
sns.jointplot(df1['day'],df1['sex'])
```

```python
### it returns the kind paramter plot as margins and main plot as histogram
sns.jointplot(df1['total_bill'],df1['tip'],kind = 'kde')
```

```python
sns.jointplot(df1['total_bill'],df1['tip'],kind = 'scatter')
```

```python
sns.jointplot(df1['total_bill'],df1['tip'],kind = 'reg')
```

```python
sns.jointplot(df1['total_bill'],df1['tip'],kind = 'hex')
```
