### 6. Scaling the features in the same range

- Standard Scaler  ->mean =0,std=1
- minmax Scaler -> [0,1]
- robust Scaler -> iQrange
- normalizer -> row wise,[0,1]

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer

s=StandardScaler()
m=MinMaxScaler()
r=RobustScaler()
n=Normalizer()
```

```python
s1 = s.fit_transform(df)  # StandardScaler

m1 = m.fit_transform(df)  # MinMaScaler

r1 = r.fit_transform(df)  # RobustScaler

n1 = n.fit_transform(df)  # Normalization
```

```python
plt.boxplot(df)
plt.title("before Scaling")
plt.show()
```

```python
plt.boxplot(s1)
plt.title("After applying Standard Scaling")
plt.show()
```

```python
plt.boxplot(m1)
plt.title("After applying MinMax Scaling")
plt.show()
```

```python
plt.boxplot(r1)
plt.title("After Applying Robust Scaling")
plt.show()
```

```python
plt.boxplot(n1)
plt.title("After applying Normalized Scaling")
plt.show()
```
