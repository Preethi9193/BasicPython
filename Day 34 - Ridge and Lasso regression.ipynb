### Regularization( Ridge and Lasso Regression)

- It is one of Tuning technique
- It is used to slove the problem of overfitting and also 
- used to reduce the errors by fitting the function approriately on the training data and avoid the overfitting
- slightly do some modifications to the learning algorithm
- Result is it improves the model performance on the unseen(test) data
#### Techniques are

- 3 techniques 
    - Ridge
    - Lasso
    - Elastic net

#####  Overfitting 
- training data is less ,test data is more -> performance  will be low
- it can be overcome by using ridge and lasso regression

### Ridge regression(L2 Regression)

- when their is a unit change in x may varies huge change in y(slope) which comes under a overfitting condition . so for decreasing those huge increasing in slope,we choose another best fit line which adds some error.
- we can choose best fit lines by taking lambda value from 1 to  any no of values.

- Here we choose in result which is having the lowest error then we consider that lambda value.

- Performance will be good on test data also
- costfunction + (lambda * (slope)^2)

```python
## to import datasets from the sklearn

#from sklearn import datasets
from sklearn.datasets import load_boston
```

```python
## to load the dataset
df = load_boston()
df
```

```python
## to get only keys in dataset
df.keys()
```

```python
## to create a dataframe

boston = pd.DataFrame(df.data)
```

```python
boston.columns
```

```python
## to add columns names to dataset
boston.columns = df.feature_names
```

```python
boston.head()
```

```python
## to add any column in dataset and values from dataset
boston["Target"] = df.target
boston
```

```python
## to check whether it has any cleaning
boston.isnull().sum()
```

```python
boston.info()
```

```python
### to select input and output values

x = boston[boston.columns[:-1]]   # to take all columns as input except one column
y = boston['Target']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
x_scale = scale.fit_transform(x)
```

```python
x_scale
```

```python
## to apply model

from sklearn.linear_model import Ridge

## to split the data into train and test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_scale,y,train_size = 0.7,random_state = 25)

model = Ridge(alpha = 1)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_test,y_test)*100
```

```python
model.score(x_train,y_train)*100
```

### Lasso regression (L1 Regularization)

- identical to ridge(reduces the overfitting)
- costfunction + lambda * |slope| ---> formula
- another usage of lasso is removes unwanted features(magintudes of the unwanted features will be zero)
- Lasso make some values as zero beacuse it uses the maginitude

```python
from sklearn.linear_model import Lasso

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_scale,y,train_size = 0.7,random_state = 25)

model = Lasso(alpha = 0.001)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_test,y_test)*100
```

```python
model.score(x_train,y_train)*100
```
