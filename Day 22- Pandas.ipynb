### Date and time
to work with date and time formates we can use
- pd.date_range()

```python
d1 = pd.date_range('01-01-2021',periods =10)
print(d1)
```

```python
d2 = pd.date_range('01-01-2021',periods =10,freq = '1H')
print(d2)
```

```python
d1 = pd.date_range('01-01-2021','10-01-2021',periods =10)
print(d1)
```

```python
d2 = pd.date_range('01-01-2021',periods =10,freq = '24H')
print(d2)
```

```python
d2 = pd.date_range(start = '01-01-2021',end = '01-02-2021',freq = '5s')
print(d2)
```

```python
### date with series object
date = pd.Series(pd.date_range(start = '01-01-2021',end = '01-02-2021',freq = '24h'))
print(date)
print(type(date))
```

## to create data frame   -> pd.DataFrame()
- similar to 2d arrays
- by using list,tuple,numpy array, dict formats  we can create 
- when we have string,float,complex -> we can create as string/object

```python
df = pd.DataFrame([[1,2,3],[4,5,'preeti']])  # using list
print("using list",df)

df1 = pd.DataFrame((1,3),(1,2))  # using tuple  ---- here second place value consider as index
print("using tuple",df1) 

df2 = pd.DataFrame({'a':[1,4],'b':[2,5],'c':[3,6]})   # using dict
print("using dict",df2)

df3 = pd.DataFrame(np.array([[1,3,4],[2,5,6]]))  # using numpy array
print("using numpy array",df3)
```
