### Concatenation of data frames

- pd.concatenation(df1,df2)

```python
d1 = pd.DataFrame({'clg':['MIC','Gist','VVIT','SRM'],
                   'Temp':[32,63,45,31],
                    'Hum':[20,46,57,13]})
d1
```

```python
d2 = pd.DataFrame({'clg':['Abc','Kits','nec','xyz'],
                  'Temp':[40,35,67,89],
                  'Hum':[50,87,90,65]})
d2
```

```python
d3 = pd.concat((d1,d2))
```

```python
d3
```

```python
# d2 = pd.DataFrame({'clg1':['Abc','Kits','nec','xyz'],
#                  'Temp1':[40,35,67,89],
#                  'Hum1':[50,87,90,65]})
#d2
```

```python
d4 = pd.concat((d1,d2))
d4
```

```python
d3 = pd.concat((d1,d2),ignore_index = True)    ### to get index in line
d3
```

```python
## To specific different dataframe we can use keys and values to it 
## then just ignore index_ignore = true

d3 = pd.concat((d1,d2),keys = ['k1','k2']) 
d3
```

### merging data frames
- pd.merge()
- while using cols name sholud be same atleast one and values of some
- when we merge it will return only common values


```python
d2
```

```python
a = pd.DataFrame({'x':[10,20,30,40],'y':[30,40,50,60]})
b = pd.DataFrame({'u':[1,2,3,4],'y':[30,20,50,60]})
```

```python
a
```

```python
b
```

```python
c = pd.merge(a,b,indicator = True )
c
```

```python
c = pd.merge(a,b)
c
```

```python
c = pd.merge(a,b,how = 'right',indicator = True)
c
```

```python
c = pd.merge(a,b,on = 'y',how = 'right',indicator = True)
c
```
