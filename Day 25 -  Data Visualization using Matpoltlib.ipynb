### Scatter plot

- used in Machine Learning popularly
- used for identifying the relationship between two features
- +ve
- -ve
- neutral

```python
x = [1,2,3,4,5]
y = np.array([2,4,6,8,10])
y2 = -y
y3 = [5,5,5,5,5]
```

```python
plt.scatter(x,y)   ## +ve scattter
plt.show()
```

```python
plt.scatter(x, y2)   ## -ve scatter
plt.show()
```

```python
plt.scatter(x, y3)   ## neutral scatter
plt.show()
```

```python
plt.scatter(x, y,c= ['b','r','g','k','#ffaabb'], alpha = 0.9)
plt.show()
```

```python
plt.scatter(x, y,marker = '>',c = ['b','r','g','k','#ffaabb'], alpha = 0.9)
plt.show()
```

```python
plt.scatter(x, y,marker ='*',c= ['b','r','g','k','#ffaabb'], alpha = 0.9, s = y **3)
plt.show()
```

```python
plt.figure(figsize=(10,15))
plt.scatter(x, y,marker ='*',c= ['b','r','g','k','#ffaabb'], alpha = 0.9, s = y **3, edgecolors = 'k')
plt.show()
```
