### Filtering a dataframe

accessing a particular data or info from given data frame

```python
dfcsv['temparature']
```

```python
dfcsv.loc[(dfcsv.temparature > 55)]   ## filtering data using loc()
```

```python
 ### filtering data using data[data(condition)]
dfcsv[dfcsv['temparature'] > 55]  
```

```python
dfcsv
```

```python
## getting data of winspeed data above 13

dfcsv[dfcsv['winspeed'] > 13.0]
```

```python
## filter data temp > 40 and winspeed >15

dfcsv[(dfcsv['temparature'] >= 30) & (dfcsv['winspeed'] == 15)]   ## both cond true only
```

```python
 ## filter data temp > 40 or winspeed >15

dfcsv[(dfcsv['temparature'] >= 30) | (dfcsv['winspeed'] == 15)]   ## 1 con true 
```

```python
## to get count of null values in each column
### isnull() will return the nan present

dfcsv.isnull().sum()
```

```python
### to remove nan from dataset we use dropna() it remove total row from dataset

d1 = dfcsv.dropna()
d1
```

```python
d1.isnull().sum()
```

```python
###

dfcsv.notnull().sum()
```

```python
### to handle null values we can fill the data by fillna(val)

dfcsv.fillna(0)
```

```python
## to fill according to columnwise
 
dfcsv.fillna({'temparature':1,'winspeed':2})
```

```python
## to fill according to columnwise and in main dataframe we use inplace
 
dfcsv.fillna({'temparature':1,'winspeed':2},inplace = True)
```

```python
## to fill according to columnwise and values
 
dfcsv.fillna({'temparature':1,'winspeed':2,'event':'sunny'})
```
