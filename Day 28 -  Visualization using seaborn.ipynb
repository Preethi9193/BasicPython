### Creation of subplots

- plt.subplot(row,col,index/position) -> to create a subplot 
- subplot generate only single row
- index sholud be less than row and col size

```python
## to create a single subplot
plt.subplot(1,2,1)   # subplot(row,cols,index/position)
```

```python
plt.subplot(1,2,2)    
```

```python
## when we need to plot in single subplot
plt.subplot(1,2,1)
plt.plot([1,2,3],[2,3,4])
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[2,3,4])
plt.subplot(1,2,2)
plt.scatter([1,2,3,4,5],[5,6,7,8,9])
```

```python
plt.subplot(2,1,1)
plt.subplot(2,1,2)
```

```python
plt.subplot(2,1,1)   # two rows,single col and index is 1
plt.bar(['x','y','z'],height = 5)

plt.subplot(2,1,2)   # two rows,single col  and index is 2
plt.hist(np.random.randint(1,25,10),bins = 5)

#plt.subplot(2,1,3)   # it return error because row has 2 but we gave index is 3
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,7,10,15])
plt.subplots(2,2)    # to get more plots at a time subplots 
## -- here wee get 2 rows and 2 col plots
plt.subplot(2,2,1)
plt.plot(x,x,'r*',label = 'Line plot')
plt.legend()
plt.subplot(2,2,2)
plt.scatter(x,y,label = 'Scatter plot')
plt.legend(loc='upper left')
plt.subplot(2,2,3)
plt.bar(x,y,color ='k',label = 'Bar plot')
plt.legend()
plt.subplot(2,2,4)
plt.barh(x,y,color ='green',label = 'BarHorizontal plot')
plt.legend()  # it shows label name on the plot 
```

```python
plt.subplot(1,2,1)
plt.pie([10,20,30],labels =['python','java','R'],labeldistance =0.5)
plt.subplot(1,2,2)
plt.barh([10,15,20],width = 5,color= ['r','m','b'])
plt.text(1,10,'m1')
plt.text(1,15,'m2')
plt.text(2,20,'m3')
```

```python
### to create subplot another method add_subplot 
### but figure need to be used then only we can use add_subplot() 

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

```python
fig = plt.figure()

### to create line plot
ax1 = fig.add_subplot(2,2,1)
plt.plot(np.random.randn(50).cumsum(),'g',label = 'Line plot')
plt.legend()

### to create histogram  -> to get horizontally we use orientation
ax2 = fig.add_subplot(2,2,2)
ax2.hist(np.random.randn(100),bins = 20,color = 'grey',orientation= 'horizontal',label = 'Histogram')
plt.legend()

### to create scatter plot
ax3 = fig.add_subplot(2,2,3)  
ax3.scatter(np.arange(30),np.random.randint(1,50,30),color = 'red',label = 'Scatter plot')
plt.legend(loc = 'upper left')

plt.show()
```

```python
### to get the possibilties of colors in seaborn

sns.color_palette()
```

```python
### to get color after running sns.color_palette cell to get output at outside also we use sns.palplot()
### palplot -> will display a color's palette 
### (it is like object which is combination of colors )
sns.palplot(sns.color_palette())
```

```python
### to generate fixed noof color as output

sns.palplot(sns.color_palette(n_colors = 8))
```

```python
### To get only presented colors 

sns.palplot(sns.color_palette(n_colors = 15,as_cmap = True))
```

```python
### to generate own color

sns.palplot(sns.color_palette("PuBu"))
```

```python
sns.palplot(sns.color_palette("coolwarm"))
```

```python
### To apply styles  -- and it takes only white, dark, whitegrid, darkgrid, ticks

sns.set_style('darkgrid')
sns.palplot(sns.color_palette("coolwarm"))
```

```python
### to get sequential colors -- start Low range to high range

sns.cubehelix_palette()
```

```python
sns.palplot(sns.cubehelix_palette())
```

```python
sns.palplot(sns.cubehelix_palette(n_colors= 10)) 
```

```python
### to get reverse order

sns.palplot(sns.cubehelix_palette(reverse = True))
```

```python
## to control intensity of color

sns.palplot(sns.cubehelix_palette(reverse = True,hue = 1))
```

```python
### to create own custom palette

sns.palplot(sns.light_palette('blue'))   # it generate low to dark color 
```

```python
sns.palplot(sns.light_palette('blue',n_colors =7,reverse = True))
```

```python
### to create dark own custom color
    
sns.palplot(sns.dark_palette('blue'))   ### generate dark to low
```

```python
sns.palplot(sns.dark_palette('blue',n_colors =7,reverse = True))
```
